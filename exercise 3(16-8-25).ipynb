```python
with open('new.txt','w')as file:
  file.write("hello")
  file.write("\n hyy")
  file.write("\n haaaaaai")
with open('new.txt','r')as file:
  a=file.read()
  print(a)
with open('new.txt','r')as file:
  a1=file.readline()
  print(a1)
with open('new.txt','r')as file:
  a2=file.readlines()
  print(a2)
```
