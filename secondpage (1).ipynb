**Bank account**

```python
class BankAccount:
  def __init__(self,owner,balance=0):
    self.owner=owner
    self.balance=balance
  def deposit(self,amount):
    self.balance+=amount
    print(f"{amount} is deposited.New balance is{self.balance}")
  def withdraw(self,amount):
    if amount>self.balance:
      print("Insufficient funds")
    else:
      self.balance-=amount
      print(f"{amount}is withdrawn.New Blance is {self.balance}")
  def get_balance(self):
    return self.balance

act=BankAccount("Krish",500000)
print(act.owner)
print(act.balance)
print(act.deposit(600000))
print(act.withdraw(100000))
print(act.get_balance())
act.deposit=10
```

Single Inheritance

```python
class Parent:  #parent class#
  def greet(self):
    print("Hi Parent")
class Child(Parent):  #child class#
  def greet_c(self):
    print("Hi Child")
c=Child()
c.greet() #calling function of parent class from child class#
c.greet_c() #calling function of child class from child class#
```

**Multiple Inheritance**

```python
class A:  #Parent class 1 #
  def one(self):
    print("Hi A")
class B: #Parent class 2 #
  def two(self):
    print("Hi B")
class C(A,B): #Child class#
  def three(self):
    print("Hi C")
obj=C()
obj.one()
obj.two()
obj.three()
```

**Multilevel Inheritance**

```python
class One: #grandparent#
  def a(self):
    print("Hi")
class Two(One): #parent#
  def b(self):
    print("Hello")
class Three(Two): #child#
  def c(self):
    print("Haaaii")

ob=Three()
ob.a()
ob.b()
ob.c()
```

```python
class One: #grandparent#
  def a(self):
    print("Hi")
obj=One()
obj.b()
class Two(One): #parent#
  def b(self):
    print("Hello")
class Three(Two): #child#
  def c(self):
    print("Haaaii")

ob=Three()
ob.a()
ob.b()
ob.c()
```

**Heirarchial Inheritance**

```python
class Aa: #parent class#
  def a1(self):
    print("Hii")

class BB(Aa): #child class 1 #
  def b1(self):
    print("Hello")
ob1=BB()
ob1.a1()
ob1.b1()

class CC(Aa): #child class 2 #
  def c1(self):
    print("Haai")
ob2=CC()
ob2.a1()
ob2.c1()
```

**Hybrid Inheritance**

```python
class A: #parent 1#
  def a1(self):             #heirarchial#
    print("1")

class B(A): #child of A#
  def b1(self):
    print("2")
ob=B()
ob.a1()
ob.b1()

class C(A): #child of A#
  def c1(self):
    print("3")
ob1=C()
ob1.a1()
ob1.c1()

class G(B,C): #child of A and B , bcz B inherited A#
  def d1(self):
    print("4")             #multiple#
ob2=G()
ob2.a1()
ob2.b1()
ob2.c1()
ob2.d1()
```

**library**

```python
class library:
    def __init__(self,title,author,stock):
        self.title=title
        self.author=author
        self.stock=stock
class borrow(library):
    def borrow(self,title,author,n):
        if self.stock<0:
            print("not available")
        else:
            self.stock-=n
            print(f"{title} by {author} is borrowed . Available stock is {self.stock}")
class retur(borrow):
    def returned(self,title,author,n):
        self.stock+=n
        print(f"{title} by {author} is returned . Available stock is {self.stock}")
obj=retur("h","k",3)
obj.returned("abc","him",10)
obj.borrow("def","her",3)
```
